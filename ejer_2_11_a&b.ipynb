```
def factorial(n):
  i=1
  fac=1
  while i<=n:
    fac=fac*i 
    i=i+1
  return fac
```

```
def binomial(n,k):
  if k > 0:
    res=factorial(n)/(factorial(k)*factorial(n-k))
  elif k==0:
    res=1
  elif k < 0:
    return(print("k<0, es imposoble de calcular"))
  return(int(res))
```
